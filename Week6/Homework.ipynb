```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/datasets/master/housing.csv')
```

```python
df
```

```python
columns = ['latitude',

'longitude',

'housing_median_age',

'total_rooms',

'total_bedrooms',

'population',

'households',

'median_income',

'median_house_value',

'ocean_proximity']
```

# Loading the data


```python
df = df.fillna(0)
df["median_house_value"] = df["median_house_value"].apply(lambda x: np.log(x))
```

```python
df_train, df_val = train_test_split(df, test_size=0.4, random_state=1)
df_val, df_test = train_test_split(df_val, test_size=0.5, random_state=1)
```

```python

from sklearn.feature_extraction import DictVectorizer

train_dicts = df_train[columns].to_dict(orient='records')
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)

X_val = dv.transform(df_val[columns].to_dict(orient='records'))
```

```python
X_train
```

# Question 1


```python
# Train regressor using Xgboost
import xgboost as xgb

y_train = df_train['median_house_value'].values
y_val = df_val['median_house_value'].values

dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)

params = {
    'max_depth': 1,
}

model = xgb.train(params, dtrain, num_boost_round=100, evals=[(dval, 'val')])
model
```
